WAP in python that calculates the grade/score a student based on number of hours studied

```python
import pandas as pd
import numpy as np
```

```python
dataset= pd.read_excel("/content/lab1.xlsx")
dataset.head(3)
```

```python
x_mean = np.mean(dataset['Studied'])
x_mean
```

```python
y_mean = np.mean(dataset['Marks'])
y_mean
```

```python
dataset['diff_x'] = dataset['Studied'] - x_mean
dataset['diff_y'] = dataset['Marks'] - y_mean
dataset.head(3)
```

```python
dataset['mul_xy'] = dataset['diff_x'] * dataset['diff_y']
dataset.head(3)
```

```python
dataset['square_diff_x'] = dataset['diff_x'] * dataset['diff_x']
dataset.head(3)
```

```python
numerator = np.sum(dataset['mul_xy'])
denominator = np.sum(dataset['square_diff_x'])
```

```python
x_mean = np.mean(dataset['Studied'])
y_mean = np.mean(dataset['Marks'])
```

```python
m = numerator / denominator
c = y_mean - (m * x_mean)
```

```python
print(m)
print(c)
```

```python
x = int(input("enter the number of hours studied"))
y = m*x+c
print("The grade that you may get bases upon number of hours you studied is : ", y)
```
